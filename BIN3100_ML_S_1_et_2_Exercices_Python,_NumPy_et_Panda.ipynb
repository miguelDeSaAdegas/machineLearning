Ecrivez une fonction qui va générer une matrice de nombres réels qui aura n lignes et m+1 colonnes. Dans les m premières colonnes les nombres seront choisis aléatoirement selon une distribution normale et la dernière colonne sera remplie de 1.

```python
import numpy as np
```

```python
def initialisation(n,m):
  # n : nombre de lignes
  # m : nombre de colonnes
  # renvoie une matrice n x (m+1) avec des nombres aléatoires (selon une distribution normale) dans les m premières colonnes
  # et des 1 dans la dernière colonne
  # A COMPLÉTER
  array_1 = np.random.randn(n,m)
  array_2 = np.ones((n,1))
  array_3 = np.concatenate((array_1, array_2), axis = 1, dtype = np.float16)

  return array_3

print(initialisation(3,3))
```

Soit p un entier et M un ndarray.

Ecrivez une fonction zoom(M,p) qui renvoie un ndarray qui est la sous-matrice contenant les éléments se trouvant dans les p % centraux de la matrice M.

Par exemple si la matrice de départ à 20 lignes et 20 colonnes (400 éléments)
et que l'on demande les 25% centraux.

Il faudra renvoyer la sous-matrice 10x10 (100 éléments) se trouvant au centre de la matrice. Donc les éléments aux indices de lignes et de colonnes entre 5 et 15.

```python
def zoom(M,p):
  #M : la matrice sur laquelle on va "zoomer"
  #p : le pourcentage central de la matrice à renvoyer
  #A compléter
  horizontale = M.shape[1]
  verticale = M.shape[0]

  reduHor = int((horizontale/100)*(p/2))
  reduVert = int((verticale/100)*(p/2))

  newPhoto = M[(0+reduHor):((horizontale-reduHor)+1),(0+reduVert):((verticale-reduVert)+1)]

  return newPhoto
```

1. Prenez le fichier BTC-EUR.csv afin d'importer les données du Bitcoin, en les indexant par date, dans un dataFrame panda que l'on appelera bitcoins.
2. Ajoutez une colonne 'rollingMin' contenant la valeur du minimale du bitcoin sur les 28 derniers jours (en faisant un shift de 1 !)
3. Ajoutez une colonne 'rollingMax' contenant la valeur du maximale du bitcoin sur les 28 derniers jours (en faisant un shift de 1 !)
4. Ajoutez une colonne 'sell' qui contient -1 si la valeur de clôture du jour est strictement plus petite que le minimum des 28 derniers jours et 0 sinon
5. Ajoutez une colonne 'buy' qui contient 1 si la valeur de clôture jour est strictement plus grande que le maximum des 28 derniers jours et 0 sinon
6. (bonus) Sur une seule figure, tracez deux graphiques en ne prenant que les données de 2019 : un graphique avec les valeurs de clôture, minimal et maximal sur 28 jours et un graphique avec le temps en absisse et 'buy' et 'sell' en ordonnée.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

btc = pd.read_csv('BTC-EUR.csv', index_col='Date', parse_dates=True)

btc['rollingMin'] = btc.loc['2019', 'Close'].rolling(window = 28).min().shift(1)
btc['rollingMax'] = btc.loc['2019', 'Close'].rolling(window = 28).max().shift(1)

btc['sell'] = np.zeros(len(btc))
btc['buy'] = np.zeros(len(btc))

btc.loc[btc['Close'] < btc['rollingMin'],'sell'] = -1
btc.loc[btc['Close'] > btc['rollingMax'],'buy'] = 1

start = '2019'
end = '2019'
fig, ax = plt.subplots(2, figsize=(12, 8), sharex = True)
ax[0].plot(btc['Close'][start:end])
ax[0].plot(btc['rollingMin'][start:end])
ax[0].plot(btc['rollingMax'][start:end])
ax[0].legend(['close', 'min', 'max'])
ax[1].plot(btc['buy'][start:end], c='g')
ax[1].plot(btc['sell'][start:end], c='r')
ax[1].legend(['buy', 'sell'])
```
